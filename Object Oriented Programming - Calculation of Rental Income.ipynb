```python
class ROICalculator():
    
    incomes = 0
    expenses = 0
    investments = 0
    
    '''
    This calculator class will have income, expenses, and investments. 
    It will have three methods, take income, take expenses, take investments, 
    and calculate the CashonCash ROI.
    '''
    
    def __init__(self, incomes=0, expenses=0, investments=0):
        self.incomes = incomes
        self.expenses = expenses
        self.investments = investments
    
    def take_incomes(self):
        income_input = input("\nDo you have an income to add? (y/n) \n")
        if (income_input.lower() == 'y'):
            income = input('What is the value of the income? \n')
            try:
                income = int(income)
                self.incomes += income
                self.take_incomes()
            except ValueError:
                print("You will need to enter a whole number.")
                self.take_incomes()
        elif (income_input.lower() == 'n'):
            print(f'Total of all incomes: {self.incomes}')
        else:
            print('Please enter a valid input')
            self.take_incomes()
    
    def take_expenses(self):
        expense_input = input("\nDo you have an expense to add? (y/n) \n")
        if (expense_input.lower() == 'y'):
            expense = input('What is the value of the expense? \n')
            try:
                expense = int(expense)
                self.expenses += expense
                self.take_expenses()
            except ValueError:
                print("You will need to enter a whole number.")
                self.take_expenses()
        elif (expense_input.lower() == 'n'):
            print(f'Total of all expenses: {self.expenses}')
        else:
            print('Please enter a valid input')
            self.take_expenses()
        
    def take_investments(self):
        inv_input = input("\nDo you have an investment to add? (y/n) \n")
        if (inv_input.lower() == 'y'):
            inv = input('What is the value of the investment? \n')
            try:
                inv = int(inv)
                self.investments += inv
                self.take_investments()
            except ValueError:
                print("You will need to enter a whole number.")
                self.take_investments()
        elif (inv_input.lower() == 'n'):
            print(f'Total of all investments: {self.investments}')
        else:
            print('Please enter a valid input')
            self.take_investments()
        
    def calculate_roi(self):
        if (self.incomes == 0):
            print("Either you didn't enter some incomes or this isn't a good investment.")
            self.take_incomes()
        elif (self.expenses == 0):
            print("Either you didn't enter some expenses or this is a steal!") 
            self.take_expenses()
        elif (self.investments == 0):
            print('Investments cannot equal zero. Please enter some investment values.')
            self.take_investments()
        else:
            print(f'\n \nThe CashonCash ROI for this property is {100*(self.incomes - self.expenses)*12/self.investments}%\n \n')
            return

newROICalculator = ROICalculator()
```
